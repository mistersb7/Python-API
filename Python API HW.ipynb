```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
# Import API key
#from api_keys import api_key
from api_keys import api_key


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
cities
```

```python
#cities= ["samsu","london"]
cloudiness= []
citiesFound = []
latitude=[]
longitude=[]
country= []
date= []
humidity= []
max_temps= []
wind_speed= []
for city in cities:
    try:
        url= "http://api.openweathermap.org/data/2.5/weather?"
        query_url=url + "appid=" + api_key + "&q=" + city + "&units=imperial"
        weather_response= requests.get(query_url).json()
        citys= weather_response["name"]
        citiesFound.append(citys)
        lats= weather_response["coord"]["lat"]
        latitude.append(lats)
        longs=weather_response["coord"]["lon"]
        longitude.append(longs)
        cloudy= weather_response["clouds"]["all"]
        cloudiness.append(cloudy)
        ctry=weather_response["sys"]["country"]
        country.append(ctry)
        dates= weather_response["dt"]
        date.append(dates)
        humid= weather_response["main"]["humidity"]
        humidity.append(humid)
        maxtemp=weather_response["main"]["temp_max"]
        max_temps.append(maxtemp)
        windspeed=weather_response["wind"]["speed"]
        wind_speed.append(windspeed)
        print(f"Processing Record {cities.index(city)+1} | {city}")
    except(KeyError):
        print("City not found. Skipping...")
```

```python
weather_dict={
    "City": citiesFound,
    "Lng": longitude,
    "Lat": latitude,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Max Temp": max_temps,
    "Wind Speed": wind_speed
}

weatherdf= pd.DataFrame(weather_dict)
weatherdf.to_csv("city_weather.csv")
weatherdf.count()
```

```python
from time import gmtime, strftime
timestamp = weatherdf.Date[1]
converted_date= strftime("%x",gmtime(timestamp))
```

```python
plt.scatter(weatherdf["Lat"],weatherdf["Max Temp"], marker="o", alpha= .75, edgecolors="black")
plt.title(f"City Latitude vs. Max Temperature ({converted_date})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature(F)")
plt.grid(True)
plt.savefig("Lattitude_vs_MaxTemperature.png")
plt.show()
```

```python
plt.scatter(weatherdf["Lat"],weatherdf["Humidity"], marker="o", alpha= .75, edgecolors="black")
plt.title(f"City Latitude vs. Humidity ({converted_date})")
plt.xlabel("Latitude")
plt.ylabel("Humidity(%)")
plt.ylim(0,105)
plt.grid(True)
plt.savefig("Lattitude_vs_Humidity.png")
plt.show()
```

```python
plt.scatter(weatherdf["Lat"],weatherdf["Cloudiness"], marker="o", alpha= .75, edgecolors="black")
plt.title(f"City Latitude vs. Cloudiness ({converted_date})")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness(%)")
plt.ylim(0,105)
plt.grid(True)
plt.savefig("Lattitude_vs_Cloudiness.png")
plt.show()
```

```python
plt.scatter(weatherdf["Lat"],weatherdf["Wind Speed"], marker="o", alpha= .75, edgecolors="black")
plt.title(f"City Latitude vs. Wind Speed ({converted_date})")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed(mph)")
plt.ylim(0,weatherdf["Wind Speed"].max()+5)
plt.grid(True)
plt.savefig("Lattitude_vs_WindSpeed.png")
plt.show()
```

Observations:
1- There is a correlation between a city's latitude and it's max temperature. The further away from 0 latitude, the lower the experienced max temperature.
2- There is a slight propensity for higher average humidity near the equator. A larger sample set may allow us to further prove the hypothesis.
3- Based on the sample data set, it is unlikely to experience wind speeds higher than 30 mph. 
4- The sample set is slighty skewed towards the northern hemisphere, but not subtantially more.
